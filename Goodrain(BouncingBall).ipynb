```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gravity = 4.905
diameter = 0.1
mass = 0.5
height = 5
velocity = 0
netForce = gravity
data = []
time = 0
while time < 1.5:
    netForce = gravity
    acc = netForce / mass
    velocity = velocity + acc * 0.001
    if velocity > 0:
        print("Positive Velocity")
    else:
        print("Negative Velocity")
    height = height - velocity * 0.001
    if 2.5-0.01 < height < 2.5+0.01:
        print(time)
    data.append(height)
    if data[-1] < diameter/2:
        netForce = gravity - 100
        acc = netForce / mass
        velocity = velocity + acc * 0.001
        height = height - velocity * 0.001
        data.append(height)
    time += 0.001
       
x = np.arange(0,len(data)*0.01,0.01)
plt.figure(
    figsize=(10,4),
    dpi=80
)
plt.plot(
    x,
    data,
    color='blue',
    markersize=12,
    markeredgecolor='g',
)
plt.title("Height v.s Time(Fall)")
plt.xlabel("Time(s)")
plt.ylabel("Height")
plt.show()
```
