```python
drag_force = float(input("Please input a drag force(unit:N)"))
# F = ma
gravity = 4.905
mass = 0.5
if gravity > drag_force:
    total_force = gravity - drag_force
elif gravity < drag_force:
    total_force = drag_force - gravity
acc = total_force / mass
print(f"acc is {acc} m/s^2")
```

```python
def predict(drag_force: float) -> float:
    gravity = 4.905
    mass = 0.5
    if gravity > drag_force:
        total_force = gravity - drag_force
    elif gravity < drag_force:
        total_force = drag_force - gravity
    acc = total_force / mass
    return acc

predict(2)
```

### Homework: Drop Problem

```python
gravity = 4.905
mass = 0.5
height = 5
velocity = 0
acc = gravity / mass
data = []
for i in range(0,50):
    velocity = velocity + acc * 0.01
    height = height - velocity * 0.01
    data.append(height)

print(data)
print(height)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,0.5,0.01)
plt.figure(
    figsize=(10,4),
    dpi=80
)
plt.plot(
    x,
    data,
    marker="_",
    color='blue',
    markersize=12,
    markeredgecolor='g',
)
plt.title("Height v.s Time(Fall)")
plt.xlabel("Time(s)")
plt.ylabel("Height")
plt.show()
```
