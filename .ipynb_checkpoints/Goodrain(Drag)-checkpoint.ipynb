```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diameter = 0.1 #Unit m
h = 0.5 #Unit m
mass = 10 #Unit Kg
g = 9.8 #n/g
gravity = mass * g

b = 0.5

height = 10
velocity = 0
netForce = gravity
data = []
time = 0
while height > h:
    dragForce = b * velocity
    netForce = gravity - dragForce
    acc = netForce / mass
    velocity = velocity + acc * 0.001
    height = height - velocity * 0.001
    data.append(height)
    if data[-1] < diameter/2:
        netForce = gravity - 100
        acc = netForce / mass
        velocity = velocity + acc * 0.001
        height = height - velocity * 0.001
        data.append(height)
    time += 0.001
```

```python
x = np.arange(0,len(data)*0.01,0.01)
plt.figure(
    figsize=(10,4),
    dpi=80
)
plt.plot(
    x,
    data,
    color='blue',
    markersize=12,
    markeredgecolor='g',
)
plt.title("Height v.s Time(Fall)")
plt.xlabel("Time(s)")
plt.ylabel("Height")
plt.show()
```
